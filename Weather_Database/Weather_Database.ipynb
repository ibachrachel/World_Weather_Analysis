```python
# Import dependencies (initial set up)

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#3. Create a new set of 2,000 random latitudes and longitudes.

lat = np.random.uniform(low= -90.000, high= 90.000, size= 2000)
lngs = np.random.uniform(low= -180.000, high= 180.000, size= 2000)
lat_lngs = zip(lat,lngs)
lat_lngs
```

```python
# Create a list to hold the latitudes and longitudes, can be more easily manipulated this way
#coordinates object will hold our 2000 random locations

coordinates = list(lat_lngs)
    
coordinates
```

```python
#4. Get the nearest city using the citipy module
#must import this dependency
from citipy import citipy
```

```python
#4. continued. 
#Create a list to hold the cities that will be generated for the coordinates
cities = []

#Nearest city will be identified using this method. 
for coordinate in coordinates: 
    city = citipy.nearest_city(coordinate[0],coordinate[1]).city_name
    
    #Add the name of the city list, if it is unique
    if city not in cities:
        cities.append(city)
        
#Check the city list to check the length: should have a long list!
len(cities)
```

```python
#5. Perform an API call with the OpenWeatherMap. 

#must import request, key, time, and datetime to set up the progression
import requests

#Import the API Key. 
#weather_api_key = ""

from config import weather_api_key

#import time
import time
from datetime import datetime
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key 
```

```python
#create an empty list to hold the weather data
city_data = []
#Print the beginning of the locations: 
print("Beginning Data Retrieval     ")
print("-----------------------------")

#Create two counters
record_count = 1
set_count = 1 

# loop through all the cities in the list. 
for i, city in enumerate(cities):
    
    #group cities in sets of 50 for logging purposes
    #space the requests out by a time frame, stops us from getting charged for overuseage 
    
    if (i % 50 == 0 and i >=50):
        set_count +=1
        record_count = 1
        time.sleep(60)
        
      # Create endpoint URL with each city.
    city_url = url + "&q=" + city.replace(" ","+")

    # Log the URL, record, and set numbers and the city.
    print(f"Processing Record {record_count} of Set {set_count} | {city}")
    # Add 1 to the record count.
    record_count += 1
    # Run an API request for each of the cities.
    try:
        # Parse the JSON and retrieve data.
        city_weather = requests.get(city_url).json()
         # Parse out the needed data.
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_weather_description = city_weather['weather'][0]['description']
        city_country = city_weather["sys"]["country"]
        
            # Convert the date to ISO standard.
        city_date = datetime.utcfromtimestamp(city_weather["dt"]).strftime('%Y-%m-%d %H:%M:%S')
        
        
            # Append the city information into city_data list.
        city_data.append({"City": city.title(),
                              "Lat": city_lat,
                              "Lng": city_lng,
                              "Max Temp": city_max_temp,
                              "Humidity": city_humidity,
                              "Cloudiness": city_clouds,
                              'Current Description': city_weather_description,
                              "Wind Speed": city_wind,
                              "Country": city_country,
                              "Date": city_date})


    # If an error is experienced, skip the city.
    except:
        print("City not found. Skipping...")
        pass

# Indicate that Data Loading is complete.
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
```

```python
#7. Add the data to a new dataframe: add the list data to a dataframe
city_data_df = pd.DataFrame(city_data)
#check the layout of the dataframe, so you know how to adjust it
city_data_df.head()
```

```python
#7. continued. put the columns in the necessary order
new_column_order = ["City", "Country", "Lat", "Lng", "Max Temp", "Humidity", "Cloudiness", "Wind Speed", "Current Description"]
city_data_df = city_data_df[new_column_order]

#make sure that the datframe took the changes that we want it to 
city_data_df.head()
```

```python
# since the dataframe matches what we want it to look like, we can export it as a csv file and save it. save it in the folder
output_data_file = "WeatherPy_Database.csv"

#since this is data needs to also be converted to a csv file, add this here
city_data_df.to_csv(output_data_file, index_label="City_ID")

#check in excel to make sure that it looks good in the CSV file. (it does) looking good!
#double check to make sure that the API key is not exposed. 
```
